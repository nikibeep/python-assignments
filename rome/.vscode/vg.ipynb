```python
import pandas as pd # To read csv
frame = pd.read_csv('E:/vgsales.csv') # Read csv into dataframe 
frame.head(1)
```

```python
frame.columns = ['Rank','Name','Platform', 'Year','Genre' ,'Publisher', 'NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']
frame = frame.set_index('Rank')
```

```python
frame.head(10)
```

```python
frame.tail(10)
```

```python
mean1 = frame['Global_Sales'].mean()
sum1 = frame['Global_Sales'].sum()
max1 = frame['Global_Sales'].max()
min1 = frame['Global_Sales'].min()
count1 = frame['Global_Sales'].count()
median1 = frame['Global_Sales'].median() 
std1 = frame['Global_Sales'].std() 
var1 = frame['Global_Sales'].var() 
```

```python
groupby_sum1 = frame.groupby(['Publisher']).sum() 
groupby_count1 = frame.groupby(['Publisher']).count()
```

```python
print ('Mean Global_Sales: ' + str(mean1))
print ('Sum of Global_Sales: ' + str(sum1))
print ('Max Global_Sales: ' + str(max1))
print ('Min Global_Sales: ' + str(min1))
print ('Count of Global_Sales: ' + str(count1))
print ('Median Global_Sales: ' + str(median1))
print ('Std of Global_Sales: ' + str(std1))
print ('Var of Global_Sales: ' + str(var1))
```

```python
print ('Sum of values, grouped by Publisher: ' + str(groupby_sum1))
print ('Count of values, grouped by Publisher: ' + str(groupby_count1))
```
