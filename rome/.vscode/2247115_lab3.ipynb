```python
def PandemicSeason(l):
    d={1:'o',
    2:'n',
    3:'l',
    4:'i',
    5:'e',
    6:'c',
    7:'a',
    8:'s'
    }
    str1=" "
    for i in l:
        if(i in d):
            str1=str1+str(d.get(i))
            print("The String iS: ",str1)
#Adding . at the end
            print("The string after adding . at the end: ",str1+'.')
# Changing the cases of Online and Classes
            print("The Capitalized string is: ",end="")
            for i in str1:
                if i =='o' or i=='c':
                    print(i.upper(),end="")
                else:
                        print(i,end="")
                        txt = str1[::-1]
                        print("\nThe reversed String is:",(txt))
l=[1,2,3,4,2,5,6,3,7,8,8,5,8]
PandemicSeason(l)
```

```python
def getTaxandCost(a):
    return a[2]+((a[3]/100)*a[2])
def getCost(a):
    return a[2]
def discount(cost,percent):
    totalcost = cost+((percent/100)*cost)
    if totalcost>60000:
        totalcost = totalcost-((2/100)*percent)
        print("\n Cost estimated =",cost)
    else:
            print("\n No Discount is applied")
            laptops=[('Dell',5,60000,4),('Vivo',4,57000,5),('HP',4,59000,6),('Samsung',3,45000,3)]
            laptops.sort(key=getTaxandCost)
            print("\n Laptops in increasing order of tax and cost ",laptops)
            laptops.sort(reverse=True,key=getCost)
            print("\n Laptops in decreasing order of cost ",laptops)
            for i in laptops:
                print("\n--->",i[0])
                discount(i[2],i[3])
                print("------------------------------------------")
```
