```python
def AlphabetWar(string):
    l_count=0
    r_count=0
    for letter in string:
        if letter in l_dic.keys():
            l_count+=l_dic[letter]
        elif letter in r_dic.keys():
            r_count+=r_dic[letter]
    if(l_count>r_count):
        print("Left side wins! by ",l_count-r_count)
    elif(r_count>l_count):
        print("Right side wins! by ",r_count-l_count)
    else:
        print("Let's fight again!")
        
l_dic={'w':4,
       'p':3,
       'b':2,
       's':1}
r_dic={'m':4,
       'q':3,
       'd':2,
       'z':1}
print("*********** PYTHON GEEK ***********")
string=input("Enter the String of war: ")
AlphabetWar(string)
```

```python
# Importing packages
import matplotlib.pyplot as plt
import numpy as np

# Define x values and heights for the bar chart
heights = np.array([7.0, 2.0, 14.0, 35.0, 42.0, 21.0, 49.0])
x = range(len(heights))

# y value from where the horizontal line has to be drawn
y_val = 24.0

# we are splitting the heights to create a stacked bar graph by the y_val
upper_bars = np.maximum(heights - y_val, 0)
lower_bars = np.minimum(heights, y_val)

# Plotting the stacked bar graph
plt.bar(x, lower_bars, width=0.5, color='y')
plt.bar(x, upper_bars, width=0.5, color='r', bottom=lower_bars)

# Plotting the horizontal line to divide through the y_val
plt.axhline(y=y_val, color='b', linestyle=':', label='red line')

plt.show()
```

```python
def countOccurrences(str, word):
     
    # split the string by spaces in a
    a = str.split(" ")
 
    # search for pattern in a
    count = 0
    for i in range(0, len(a)):
         
        # if match found increase count
        if (word == a[i]):
           count = count + 1
            
    return count      
 
# Driver code
str="butterscotch butterscotch"
word="butterscotch"
print(countOccurrences(str, word))
```
